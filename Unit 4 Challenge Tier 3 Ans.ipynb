```python
ap_1998 = [properties_cleaned.loc['1998',borough].mean() for borough in col_names] #average prices per borough in 1998

ap_2018 = [properties_cleaned.loc['2018',borough].mean() for borough in col_names] #avg prices per borough in 2018

#new dataframe for averages in 1998 and 2018
decade_differences_indeces = {'Borough':col_names} 
decade_differences = pd.DataFrame(decade_differences_indeces).set_index('Borough')
decade_differences

decade_differences.insert(0, '1998 Average Prices', ap_1998)
decade_differences.insert(1, '2018 Average Prices', ap_2018)

decade_differences
```

```python
decade_differences['1998:2018 ratio'] = decade_differences['1998 Average Prices']/decade_differences['2018 Average Prices']


decade_differences['2018:1998 ratio'] = decade_differences['2018 Average Prices']/decade_differences['1998 Average Prices']

decade_differences
```

```python
avg_increase= decade_differences['2018:1998 ratio'].mean()
avg_increase
```
